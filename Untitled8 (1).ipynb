## 🧹 Data Preparation

Before applying any machine learning models, it is essential to prepare and clean the datasets.  
This stage ensures data consistency, accuracy, and usability for analysis.

### 🧭 Objectives of Data Preparation
1. **Handle Missing Values:** Identify and treat missing or incomplete data appropriately.  
2. **Data Type Conversion:** Convert date and numerical columns into correct formats.  
3. **Remove Unnecessary Columns:** Eliminate columns with excessive missing data or irrelevant information.  
4. **Rename Columns (if needed):** Assign meaningful names for better readability.  
5. **Feature Scaling and Encoding:** Prepare data for algorithms that require normalized numerical values or encoded categorical data.  

### ⚙️ Approach
- For the **Employee Financial Data**, we will:
  - Inspect data types and null values.
  - Drop any rows/columns with too many missing values.
  - Convert financial values to numeric types.
  - Ensure categorical variables (like department) are encoded properly.

- For the **SARB Annual Data**, we will:
  - Inspect the number of usable economic indicators.
  - Keep only relevant columns (such as GDP, inflation, and interest rate).
  - Convert the “Date” column to datetime format.
  - Drop rows with no useful economic data.

By performing these steps, we ensure that both datasets are ready for **Exploratory Data Analysis (EDA)** and subsequent modeling.


```python
# ---------------------------------------
# DATA PREPARATION
# ---------------------------------------

import pandas as pd
import numpy as np

# View basic info for both datasets
print("Employee Financial Data Overview:\n")
emp_df.info()
print("\n" + "-"*50 + "\n")
print("SARB Annual Data Overview:\n")
sarb_df.info()

# --------------------------------------------------------
# 1️⃣ EMPLOYEE FINANCIAL DATA CLEANING
# --------------------------------------------------------

# Check for missing values
print("\nMissing values in Employee dataset:\n")
print(emp_df.isna().sum())

# Drop any columns with more than 50% missing data (if any)
threshold = len(emp_df) * 0.5
emp_df = emp_df.dropna(thresh=threshold, axis=1)

# Fill remaining missing numeric values with column mean
for col in emp_df.select_dtypes(include=[np.number]).columns:
    emp_df[col].fillna(emp_df[col].mean(), inplace=True)

# Check unique departments or roles (for encoding later)
print("\nUnique Departments:\n", emp_df['Department'].unique() if 'Department' in emp_df.columns else "Not Found")

# --------------------------------------------------------
# 2️⃣ SARB DATA CLEANING
# --------------------------------------------------------

# Drop columns with all missing values
sarb_df = sarb_df.dropna(axis=1, how='all')

# Keep only numeric columns that have at least some data
sarb_df = sarb_df.select_dtypes(include=[np.number]).dropna(axis=1, how='all')

# Reset index and ensure 'Date' is datetime
if 'Date' in sarb_df.columns:
    sarb_df['Date'] = pd.to_datetime(sarb_df['Date'], errors='coerce')
else:
    print("⚠️ 'Date' column not found; please check dataset structure.")

# Display shape and head of cleaned data
print("\n✅ Cleaned Employee Dataset Shape:", emp_df.shape)
print("✅ Cleaned SARB Dataset Shape:", sarb_df.shape)

print("\nCleaned Employee Data Preview:")
display(emp_df.head())

print("\nCleaned SARB Data Preview:")
display(sarb_df.head())
```

## ⚙️ Feature Engineering

Feature Engineering is the process of creating new variables (features) from existing data to improve the performance of machine learning models.

### 🎯 Objectives
1. **Enhance Predictive Power:** Derive ratios and indicators that capture financial behavior.  
2. **Simplify Relationships:** Make patterns more obvious to ML algorithms.  
3. **Reduce Noise:** Eliminate redundant or less useful features.  

### ✨ Planned Features (Employee Data)
- **Debt-to-Income Ratio:** `(Expenses / Income)` → Measures financial pressure.  
- **Savings Rate:** `(Savings / Income)` → Indicates financial discipline.  
- **Net Disposable Income:** `(Income - Expenses - Deductions)` → Amount left after essential costs.  
- **High Withdrawal Risk Flag (optional):** Rule-based variable for initial labeling or sanity check.


```python
# ---------------------------------------
# FEATURE ENGINEERING
# ---------------------------------------

# --------------------------------------------------------
# EMPLOYEE DATA - NEW FEATURES
# --------------------------------------------------------

if 'Income' in emp_df.columns and 'Expenses' in emp_df.columns:
    emp_df['Debt_to_Income'] = emp_df['Expenses'] / emp_df['Income']

if 'Savings' in emp_df.columns and 'Income' in emp_df.columns:
    emp_df['Savings_Rate'] = emp_df['Savings'] / emp_df['Income']

if 'Income' in emp_df.columns and 'Expenses' in emp_df.columns and 'Deductions' in emp_df.columns:
    emp_df['Net_Disposable_Income'] = emp_df['Income'] - (emp_df['Expenses'] + emp_df['Deductions'])

# Optional: simple heuristic withdrawal risk (for sanity check, not target variable)
if 'Debt_to_Income' in emp_df.columns and 'Savings_Rate' in emp_df.columns:
    emp_df['Risk_Flag'] = emp_df.apply(lambda row: 1 if (row['Debt_to_Income'] > 0.6 and row['Savings_Rate'] < 0.1) else 0, axis=1)

print("✅ Employee data after feature engineering:")
display(emp_df.head())

# --------------------------------------------------------
# SARB DATA - BASIC PREP
# --------------------------------------------------------

# Keep only numeric columns (besides time)
sarb_features = sarb_df.select_dtypes(include='number')

# Normalize features (optional, useful for ML)
sarb_features_norm = (sarb_features - sarb_features.min()) / (sarb_features.max() - sarb_features.min())

print("\n✅ SARB data after feature preparation (normalized sample):")
display(sarb_features_norm.head())
```

## 🤖 Model Building – Classification (Predict Withdrawals)

The first modeling task is to predict whether an employee will **withdraw from the Accessible Pot** within the next 12 months.  

### 🎯 Objective
Build a **classification model** that takes employee financial data as input and outputs a binary label:  
- `1` = Employee likely to withdraw  
- `0` = Employee not likely to withdraw  

### ⚙️ Approach
- Use the **Employee Financial Data**.  
- Select engineered features such as:  
  - `Debt_to_Income`  
  - `Savings_Rate`  
  - `Net_Disposable_Income`  
  - plus other useful columns like `Income`, `Expenses`, `Deductions`.  
- Train and evaluate diffe


```python
print("Employee Financial Data Columns:\n")
print(emp_df.columns.tolist())
```

```python
# ---------------------------------------
# MODEL BUILDING - CLASSIFICATION
# ---------------------------------------

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# --------------------------------------------------------
# 1️⃣ Define Features (X) and Target (y)
# --------------------------------------------------------

# Auto-select numeric features (ignores text columns like Name/Department)
feature_cols = emp_df.select_dtypes(include=['number']).columns.tolist()

# Remove target column if it's inside features
for col in ['Withdrawn', 'Class', 'Risk_Flag']:
    if col in feature_cols:
        feature_cols.remove(col)

X = emp_df[feature_cols].fillna(0)

# Target: Use Withdrawn if available, else Class, else Risk_Flag
if 'Withdrawn' in emp_df.columns:
    y = emp_df['Withdrawn']
elif 'Class' in emp_df.columns:
    y = emp_df['Class']
elif 'Risk_Flag' in emp_df.columns:
    y = emp_df['Risk_Flag']
else:
    raise ValueError("❌ No target variable found. Please check dataset for 'Withdrawn', 'Class', or ensure Risk_Flag was created.")

# --------------------------------------------------------
# 2️⃣ Train-Test Split
# --------------------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale features (important for Logistic Regression, not trees)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# --------------------------------------------------------
# 3️⃣ Model Training
# --------------------------------------------------------

models = {
    "Logistic Regression": LogisticRegression(max_iter=500),
    "Decision Tree": DecisionTreeClassifier(max_depth=5, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42)
}

results = {}

for name, model in models.items():
    if name == "Logistic Regression":
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

    # Evaluate metrics

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    rec = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)


    results[name] = {"Accuracy": acc, "Precision": prec, "Recall": rec, "F1": f1}

    # Print results
    print(f"\n📊 {name} Results:")
    print(f"Accuracy: {acc:.2f}, Precision (weighted): {prec:.2f}, Recall (weighted): {rec:.2f}, F1 (weighted): {f1:.2f}")



    # Confusion Matrix
    disp = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred))
    disp.plot(cmap="Blues")
    plt.title(f"Confusion Matrix - {name}")
    plt.show()

# --------------------------------------------------------
# 4️⃣ Compare Results
# --------------------------------------------------------
print("\n🔎 Model Comparison:\n")
for name, metrics in results.items():
    print(f"{name}: {metrics}")

```

## 📈 Model Evaluation

Evaluating machine learning models is crucial to measure their performance and ensure they generalize well to unseen data.

### 🎯 Goals of Evaluation
1. **Measure Accuracy:** The overall proportion of correct predictions.  
2. **Measure Precision, Recall, and F1:** Especially important for imbalanced datasets.  
   - **Precision:** How many predicted withdrawals were actually correct.  
   - **Recall:** How many actual withdrawals were detected by the model.  
   - **F1 Score:** Harmonic mean of precision and recall (balance measure).  
3. **Inspect Confusion Matrix:** To see where the model makes errors (false positives/negatives).  
4. **Evaluate Multiclass Performance:** Since our dataset has more than two classes, use **weighted precision/recall/F1** to account for class imbalance.  

### 📊 Importance
- In this project, evaluation helps determine which model (Logistic Regression, Decision Tree, Random Forest) is the most reliable for predicting employee withdrawal behavior in the **Two-Pot Retirement System**.


```python
# ---------------------------------------
# MODEL EVALUATION - DETAILED
# ---------------------------------------
from sklearn.metrics import classification_report

for name, model in models.items():
    if name == "Logistic Regression":
        y_pred = model.predict(X_test_scaled)
    else:
        y_pred = model.predict(X_test)

    print(f"\n🔎 Detailed Evaluation - {name}\n")
    print(classification_report(y_test, y_pred, zero_division=0))
```

## 📈 Forecasting Task – Locked Pot Growth (Prophet)

The second task is to **forecast the growth of the Locked Pot**, which represents the long-term savings that employees cannot access until retirement.  

### 🎯 Objective
Use **time-series forecasting** to predict future values of key economic indicators (e.g., inflation, interest rates, GDP) that influence the growth of retirement savings.

### ⚙️ Approach
- Use the **South African Reserve Bank (SARB)** dataset.  
- Select a relevant economic indicator (e.g., `GDP`, `Inflation Rate`, or `Interest Rate`) to represent Locked Pot growth trends.  
- Apply **Prophet**, a robust forecasting library that handles seasonality, trends, and missing values.  
- Forecast the indicator over the next **10 years** (or another chosen horizon).  

### 📊 Importance
Accurate forecasts help policymakers and financial planners understand how external economic conditions may affect retirement savings in the **Two-Pot System**.  
This ensures employees will have sufficient funds for retirement, even if withdrawals occur from the Accessible Pot.


```python
# ---------------------------------------
# FORECASTING WITH PROPHET
# ---------------------------------------

from prophet import Prophet
import matplotlib.pyplot as plt

# --------------------------------------------------------
# 1️⃣ Choose a numeric column to forecast
# --------------------------------------------------------
# Automatically pick the first usable numeric column (other than Year/Date)
numeric_cols = sarb_df.select_dtypes(include='number').columns
time_col = None

# Detect time column
for col in sarb_df.columns:
    if "year" in col.lower() or "date" in col.lower():
        time_col = col
        break

if not time_col:
    raise ValueError("⚠️ No Date/Year column found in SARB dataset. Please rename your year column to 'Year' or 'Date'.")

target_col = numeric_cols[0]  # pick first indicator for demo
print(f"Using {target_col} as forecasting target.")

# --------------------------------------------------------
# 2️⃣ Prepare dataset for Prophet
# --------------------------------------------------------
df_prophet = sarb_df[[time_col, target_col]].dropna()
df_prophet.columns = ['ds', 'y']   # Prophet requires these names

# --------------------------------------------------------
# 3️⃣ Build and Fit Prophet Model
# --------------------------------------------------------
model = Prophet()
model.fit(df_prophet)

# --------------------------------------------------------
# 4️⃣ Create Future DataFrame & Forecast
# --------------------------------------------------------
future = model.make_future_dataframe(periods=10, freq='Y')  # forecast 10 years ahead
forecast = model.predict(future)

# --------------------------------------------------------
# 5️⃣ Plot Forecast
# --------------------------------------------------------
fig1 = model.plot(forecast)
plt.title(f"Forecast of {target_col} (Locked Pot Growth Indicator)")
plt.show()

fig2 = model.plot_components(forecast)
plt.show()
```

- Decision Tree offered some level of non-linearity and interpretability but risked **overfitting** on training data.  
- Random Forest generally performed the best, showing **higher accuracy and balanced precision/recall**, due to combining multiple decision trees.  

### ✅ Outcome
- The **Random Forest** emerged as the most reliable model for predicting employee withdrawal behavior.  
- However, Logistic Regression remains useful when **explainability** is more important than raw accuracy.
